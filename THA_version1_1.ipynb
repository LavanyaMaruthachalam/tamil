```python
# Function to parse the content and convert it to CSV format
def parse_content_to_csv(text):
    # Split the content by double newlines to separate each kural and its explanation
    kural_blocks = text.split('\n\n')

    # Regex pattern to identify Tamil and English parts
    tamil_pattern = re.compile(r'^[\u0B80-\u0BFF]', re.MULTILINE)
    english_pattern = re.compile(r"^[A-Za-z]", re.MULTILINE)

    # Process each block and extract Tamil kural, Tamil explanation, and English translation
    rows = []
    for block in kural_blocks:
        lines = block.strip().split('\n')
        if not lines:
            continue

        tamil_lines = list(filter(tamil_pattern.match, lines))
        english_lines = list(filter(english_pattern.match, lines))

        # Assuming the first Tamil line is the kural and the rest is the explanation
        kural = tamil_lines[0] if tamil_lines else ''
        tamil_explanation = ' '.join(tamil_lines[1:]) if len(tamil_lines) > 1 else ''
        english_translation = ' '.join(english_lines) if english_lines else ''

        # Only add rows which have a kural
        if kural:
            rows.append([kural, tamil_explanation, english_translation])

    return rows

# Parse the content to structured data
csv_data = parse_content_to_csv(tamil_text_content_recent)

# Define the CSV file path
csv_output_path = '/content/sample_data/kurals.csv'

# Write the structured data to a CSV file
with open(csv_output_path, mode='w', newline='', encoding='utf-8') as csvfile:
    csvwriter = csv.writer(csvfile)
    # Writing the headers
    csvwriter.writerow(['Kural', 'Tamil Explanation', 'English Translation'])
    # Writing the data rows
    csvwriter.writerows(csv_data)

csv_output_path
```

```python
import re
import json

def parse_thirukkural(text):
    # Regular expression to capture groups based on the description
    pattern = re.compile(
        r"([^\n]+)\n([^\n]+)\n([^\n]+(?:\n[^\n]+)*?)\n([A-Z].*?)(?=\n[^\n]+\n|$)",
        re.MULTILINE | re.DOTALL
    )

    kurals = []
    matches = re.finditer(pattern, text)
    for match in matches:
        kural, tamil_explanation, extra_tamil, english_translation = match.groups()
        kurals.append({
            "kural": kural.strip(),
            "tamil_explanation": tamil_explanation.strip() + " " + extra_tamil.strip(),
            "english_translation": english_translation.strip()
        })

    return kurals

# Read the text from the file
with open('/content/drive/MyDrive/THIRUKURAL/first100.txt', 'r', encoding='utf-8') as file:
    content = file.read()

# Parse the content
parsed_kurals = parse_thirukkural(content)

# Convert to JSON
json_output = json.dumps(parsed_kurals, indent=4, ensure_ascii=False)

# Print or save the JSON data
print(json_output)
```

```python
import re

# Example string input
text = """


அகர முதல எழுத்தெல்லாம்‌ ஆதி

பகவன்‌ முதற்றே உலகு.

அகரம்‌ எழுத்துக்களுக்கு முதன்மை; ஆதிபகவன்‌, உலகில்‌ வாழும்‌ உயிர்களுக்கு முதன்மை.
'A' leads letters; the Ancient Lord

Leads and lords the entire world.

௨ கற்றதனால்‌ ஆய பயனென்கொல்‌ வாலறிவன்‌

நற்றாள்‌ தொழாஅர்‌ எனின்‌.

தன்னைவிட அறிவில்‌ மூத்த பெருந்தகையாளரின்‌ முன்னே வணங்கி நிற்கும்‌ பண்பு இல்லாவிடில்‌
என்னதான்‌ ஒருவர்‌ கற்றிருந்தாலும்‌ அதனால்‌ என்ன பயன்‌ ? ஒன்றுமில்லை

That lore is vain which does not fall

At His good feet who knoweth all.

. மலர்மிசை ஏகினான்‌ மாணடி சேர்ந்தார்‌

நிலமிசை நீடுவாழ்‌ வார்‌.

மலர்‌ போன்ற மனத்தில்‌ நிறைந்தவனைப்‌ பின்பற்றுவோரின்‌ புகழ்வாழ்வு, உலகில்‌ நெடுங்காலம்‌ நிலைத்து
நிற்கும்‌.

Long they live on earth who gain

The feet of God in florid brain.

. வேண்டுதல்வேண்‌ டாமை இலானடி சேர்ந்தார்க்கு

யாண்டும்‌ இ௫ம்பை இல.

விருப்பு வெறுப்பற்றுத்‌ தன்னலமின்றித்‌ திகழ்கின்றவரைப்‌ பின்பற்றி நடப்பவர்களுக்கு எப்போதுமே துன்பம்‌
ஏற்படுவதில்லை.

Who hold His feet who likes nor loathes

Are free from woes of human births.

இருள்சேர்‌ இருவினையும்‌ சேரா இறைவன்‌

பொருள்சேர்‌ புகழ்புரிந்தார்‌ மாட்டு.

இறைவன்‌ என்பதற்குரிய பொருளைப்‌ புரிந்து கொண்டு புகழ்‌ பெற விரும்புகிறவர்கள்‌, நன்மை தீமைகளை
ஒரே அளவில்‌ எதிர்‌ கொள்வார்கள்‌.

God's praise who tell, are free from right

And wrong, the twins of dreaming night.

பொறிவாயில்‌ ஐந்தவித்தான்‌ பொய்தீர்‌ ஒழுக்க

நெறிநின்றார்‌ நீடுவாழ்‌ வார்‌.

மெய்‌, வாய்‌, கண்‌, மூக்கு, செவி எனும்‌ ஐம்பொறிகளையும்‌ கட்டுப்படுத்திய தூயவனின்‌ உண்மையான
ஒழுக்கமுடைய நெறியைப்‌ பின்பற்றி நிற்பவர்களின்‌ புகழ்வாழ்வு நிலையானதாக அமையும்‌.

They prosper long who walk His way

Who has the senses signed away.

தனக்குவமை இல்லாதான்‌ தாள்சேர்ந்தார்க்‌ கல்லால்‌

மனக்கவலை மாற்றல்‌ அரிது.

ஒப்பாரும்‌ மிக்காருமில்லாதவனுடைய அடியொற்றி நடப்பவர்களைத்‌ தவிர, மற்றவர்களின்‌ மனக்கவலை தீர
வழியேதுமில்லை.

His feet, whose likeness none can find,

Alone can ease the anxious mind.

அறவாழி அந்தணன்‌ தாள்சேர்ந்தார்க்‌ கல்லால்‌

பிறவாழி நீந்தல்‌ அரிது.

அந்தணர்‌ என்பதற்குப்‌ பொருள்‌ சான்றோர்‌ என்பதால்‌, அறக்கடலாகவே விளங்கும்‌ அந்தச்‌ சான்றோரின்‌
அடியொற்றி நடப்பவர்க்கேயன்றி, மற்றவர்களுக்குப்‌ பிற துன்பக்‌ கடல்களைக்‌ கடப்பது என்பது எளிதான
காரியமல்ல.

Who swims the sea of vice is he

Who clasps the feet of Virtue's sea.

9. கோளில்‌ பொறியின்‌ குணமிலவே எண்குணத்தான்‌
தாளை வணங்காத்‌ தலை.
உடல்‌, கண்‌, காது, மூக்கு, வாய்‌ எனும்‌ ஐம்பொறிகள்‌ இருந்தும்‌, அவைகள்‌ இயங்காவிட்டால்‌ என்ன
நிலையோ அதே நிலைதான்‌ ஈ.டற்ற ஆற்றலும்‌ பண்பும்‌ கொண்டவனை வணங்கி நடக்காதவனின்‌
நிலையும்‌ ஆகும்‌.
Like senses stale that head is vain
Which bows not to Eight-Virtued Divine.
10. பிறவிப்‌ பெருங்கடல்‌ நீந்துவர்‌ நீந்தார்‌
இறைவன்‌ அடிசேரா தார்‌.
வாழ்க்கை எனும்‌ பெருங்கடலை நீந்திக்‌ கடக்க முனைவோர்‌, தலையானவனாக இருப்பவனின்‌ அடி
தொடர்ந்து செல்லாவிடில்‌ நீந்த முடியாமல்‌ தவிக்க நேரிடும்‌.
The sea of births they alone swim
Who clench His feet and cleave to Him.



11. வானின்‌ றுலகம்‌ வழங்கி வருதலால்‌
தானமிழ்தம்‌ என்றுணரற்‌ பாற்று.
உலகத்தை வாழ வைப்பது மழையாக அமைந்திருப்பதால்‌ அதுவே அமிழ்தம்‌ எனப்படுகிறது.
The genial rain ambrosia call
The world but lasts while rain shall fall.
12. துப்பார்க்குத்‌ துப்பாய துப்பாக்கித்‌ துப்பார்க்குத்‌
துப்பாய FTO மழை.
யாருக்கு உணவுப்‌ பொருள்களை விளைவித்துத்தர மழை பயன்படுகிறதோ, அவர்களுக்கே அந்த மழை
அவர்கள்‌ அருந்தும்‌ உணவாகவும்‌ ஆகி அரிய தியாகத்தைச்‌ செய்கிறது.
The rain begets the food we eat
And forms a food and drink concrete.
13. விண்‌இன்று பொய்ப்பின்‌ விரிநீர்‌ வியுனுலகத்து
உள்நின்‌ றூடற்றும்‌ பசி.
கடல்நீர்‌ சூழ்ந்த உலகமாயினும்‌, மழைநீர்‌ பொய்த்து விட்டால்‌ பசியின்‌ கொடுமை வாட்டி வதைக்கும்‌.
Let clouds their visits stay, and dearth
Distresses all the sea-girt earth.
14. ஏரின்‌ உழாவுர்‌ உழவர்‌ புயல்‌ என்னும்‌
வாரி வளங்குன்றிக்‌ கால்‌.
மழை என்னும்‌ வருவாய்‌ வளம்‌ குன்றிவிட்டால்‌, உழவுத்‌ தொழில்‌ குன்றி விடும்‌.
Unless the fruitful shower descend,
The ploughman's sacred toil must end.
15. கெடுப்பதாஉம்‌ கெட்டார்க்குச்‌ சார்வாய்மற்‌ றாங்கே
எடுப்பதூஉம்‌ எல்லாம்‌ மழை.
பெய்யாமல்‌ விடுத்து உயிர்களின்‌ வாழ்வைக்‌ கெடுக்கக்‌ கூடியதும்‌, பெய்வதன்‌ காரணமாக உயிர்களின்‌
நலிந்த வாழ்வுக்கு வளம்‌ சேர்ப்பதும்‌ மழையே ஆகும்‌.
Destruction it may sometimes pour
But only rain can life restore.
16. விசும்பின்‌ துளிவீழின்‌ அல்லால்மற்‌ றாங்கே
பசும்புல்‌ தலைகாண்‌ பரிது.
விண்ணிலிருந்து மழைத்துளி விழுந்தாலன்றி மண்ணில்‌ பசும்புல்‌ தலை காண்பது அரிதான ஒன்றாகும்‌.
No grassy blade its head will rear,
If from the cloud no drop appear.
17. நெடுங்கடலும்‌ தன்நீர்மை குன்றும்‌ தடிந்தெழிலி
தான்நல்கா தாகி விடின்‌.
ஆவியான கடல்நீர்‌ மேகமாகி அந்தக்‌ கடலில்‌ மழையாகப்‌ பெய்தால்தான்‌ கடல்கூட வற்றாமல்‌ இருக்கும்‌.
மனித சமுதாயத்திலிருந்து புகழுடன்‌ உயர்ந்தவர்களும்‌ அந்தச்‌ சமுதாயத்திற்கே பயன்பட்டால்தான்‌ அந்தச்‌
சமுதாயம்‌ வாழும்‌.
The ocean's wealth will waste away,
Except the cloud its stores repay.



சிறப்பொடு பூசனை செல்லாது வானம்‌

வறக்குமேல்‌ வானோர்க்கும்‌ ஈண்டு.

வானமே பொய்த்து விடும்போது, அதன்பின்னர்‌ அந்த வானத்தில்‌ வாழ்வதாகச்‌
சொல்லப்படுகிறவர்களுக்கு விழாக்கள்‌ ஏது ? வழிபாடுதான்‌ ஏது?

The earth, beneath a barren sky,

Would offerings for the gods deny.

தானம்‌ தவமிரண்டும்‌ தங்கா வியனுலகம்‌

வானம்‌ வழங்கா தெனின்‌.

இப்பேருலகில்‌ மழை பொய்த்து விடுமானால்‌ அது, பிறர்‌ பொருட்டுச்‌ செய்யும்‌ தானத்திற்கும்‌, தன்பொருட்டு
மேற்கொள்ளும்‌ நோன்புக்கும்‌ தடங்கலாகும்‌.

Were heaven above to fail below 19

Nor alms nor penance earth would show.

நீரின்‌ மையா துலகெனின்‌ யார்யார்க்கும்‌

வானின்‌ றமையா தொழுக்கு.

உலகில்‌ மழையே இல்லையென்றால்‌ ஒழுக்கமே கெடக்கூடும்‌ என்ற நிலை இருப்பதால்‌, நீரின்‌
இன்றியமையாமையை உணர்ந்து செயல்பட வேண்டும்‌.

Water is life that comes from rain 20

Sans rain our duties go in vain.




ஒழுக்கத்து நீத்தார்‌ பெருமை விழுப்பத்தூா

வேண்டும்‌ பனுவல்‌ தூணிவு.

ஒழுக்கத்தில்‌ உறுதியான துறவிகளின்‌ பெருமை, சான்றோர்‌ நூலில்‌ விருப்பமுடனும்‌, உயர்வாகவும்‌ இடம்‌
பெறும்‌.

No merit can be held so high

As theirs who sense and self deny.

தூறந்தார்‌ பெருமை துணைக்கூறின்‌ வையத்தாூ

இறந்தாரை எண்ணிக்கொண்‌ டற்று.

உலகில்‌ இறந்தவர்களின்‌ எண்ணிக்கை எவ்வளவு என்று கூற முடியுமா ? அதுபோலத்தான்‌
உண்மையாகவே பற்றுகளைத்‌ துறந்த உத்தமர்களின்‌ பெருமையையும்‌ அளவிடவே முடியாது.

To con ascetic glory here

Is to count the dead upon the sphere.

இருமை வகைதெரிந்‌ தீண்டறம்‌ பூண்டார்‌

பெருமை பிறங்கிற்‌ றுலகு.

நன்மை எது, தீமை எது என்பதை ஆய்ந்தறிந்து நன்மைகளை மேற்கொள்பவர்களே உலகில்‌
பெருமைக்குரியவர்களாவார்கள்‌.

No lustre can with theirs compare

Who know the right and virtue wear.

உரனென்னுந்‌ தோட்டியான்‌ ஓரைந்தும்‌ காப்பான்‌

வரனென்னும்‌ வைப்பிற்கோர்‌ வித்து.

உறுதியென்ற அங்குசம்‌ கொண்டு, ஐம்பொறிகளையும்‌ அடக்கிக்‌ காப்பவன்‌, துறவறம்‌ எனும்‌ நிலத்திற்கு
ஏற்ற விதையாவான்‌.

With hook of firmness to restrain

The senses five, is heaven to gain.

ஐந்தவித்தான்‌ ஆற்றல்‌ அகல்விகம்பு ளார்கோமான்‌

இந்திரனே சாலுங்‌ கரி.

புலன்களை அடக்க முடியாமல்‌ வழிதவறிச்‌ சென்றிடும்‌ மனிதனுக்குச்‌ சான்றாக இந்திரன்‌ விளங்கி,
ஐம்புலன்களால்‌ ஏற்படும்‌ ஆசைகளைக்‌ கட்டுப்படுத்தியதால்‌ வான்புகழ்‌ கொண்டவர்களின்‌ ஆற்றலை
எடுத்துக்‌ காட்டுகிறான்‌.

Indra himself has cause to say

How great the power ascetics’ sway.

செயற்கரிய செய்வார்‌ பெரியர்‌ சிறியார்‌

செயற்கரிய செய்கலா தார்‌.

பெருமை தரும்‌ செயல்களைப்‌ புரிவோரைப்‌ பெரியோர்‌ என்றும்‌, சிறுமையான செயல்களையன்றிப்‌



பெருமைக்குரிய செயல்களைச்‌ செய்யாதவர்களைச்‌ சிறியோர்‌ என்றும்‌ வரையறுத்துவிட முடியும்‌.

The small the paths of ease pursue

The great achieve things rare to do.

சவையொளி ஊறோசை நாற்றமென்‌ றைந்தின்‌

வகைதெரிவான்‌ கட்டே உலகு.

ஐம்புலன்களின்‌ இயல்பை உணர்ந்து அவற்றை அடக்கியாளும்‌ திறன்‌ கொண்டவனையே உலகம்‌ போற்றும்‌.
They gain the world, who grasp and tell

Of taste, sight, hearing, touch and smell.

நிறைமொழி மாந்தர்‌ பெருமை நிலத்து

மறைமொழி காட்டி விடும்‌.

சான்றோர்களின்‌ பெருமையை, இந்த உலகில்‌ அழியாமல்‌ நிலைத்து நிற்கும்‌ அவர்களது அறவழி நூல்களே
எடுத்துக்‌ காட்டும்‌.

Full-worded men by what they say,

Their greatness to the world display.

குணமென்னுங்‌ குன்றேறி நின்றார்‌ வெகுளி

கணமேயுங்‌ காத்தல்‌ அரிது.

குணக்குன்றுகளான பெரியவர்கள்‌ கோபம்‌ கொண்டால்‌ அந்தக்‌ கோபம்‌ அவர்கள்‌ உள்ளத்தில்‌ ஒரு கணம்‌
கூட நிலைத்து நிற்காது.

Their wrath, who've climb'd the mount of good,

Though transient, cannot be withstood.

அந்தணர்‌ என்போர்‌ அறவேோர்மற்‌ றெவ்வுயிர்க்கும்‌

செந்தண்மை பூண்டொழுக லான்‌.

அனைத்து உயிர்களிடத்திலும்‌ அன்புகொண்டு அருள்‌ பொழியும்‌ சான்றோர்‌ எவராயினும்‌ அவர்‌ அந்தணர்‌
எனப்படுவார்‌.

With gentle mercy towards all,

The sage fulfils the virtue's call.

சிறப்பினும்‌ செல்வமும்‌ ஈ.னும்‌ அறத்தினூஉங்கு

ஆக்கம்‌ எவனோ உயிர்க்கு.

சிறப்பையும்‌, செழிப்பையும்‌ தரக்கூடிய அறவழி ஒன்றைத்தவிர ஆக்கமளிக்கக்‌ கூடிய வழி வேறென்ன
இருக்கிறது ?

From virtue weal and wealth outflow;

What greater good can mankind know?

அறத்தினூஉங்‌ காக்கமும்‌ இல்லை அதனை

மறத்தலின்‌ ஊங்கில்லை கேடு.

நன்மைகளின்‌ விளைநிலமாக இருக்கும்‌ அறத்தைப்‌ போல்‌ ஒருவருடைய வாழ்க்கைக்கு ஆக்கம்‌
தரக்கூடியது எதுவுமில்லை; அந்த அறத்தை மறப்பதை விடத்‌ தீமையானதும்‌ வேறில்லை.
Virtue enhances joy and gain;

Forsaking it is fall and pain.

. ஒல்லும்‌ வகையான அறவினை ஓவாதே

செல்லும்வாய்‌ எல்லாஞ்‌ செயல்‌.

செய்யக்கூடிய செயல்கள்‌ எவை ஆயினும்‌, அவை எல்லா இடங்களிலும்‌ தொய்வில்லாத அறவழியிலேயே
செய்யப்பட வேண்டும்‌.

Perform good deeds as much you can

Always and everywhere, o man!

மனத்துக்கண்‌ மாசிலன்‌ ஆதல்‌ அனைத்தறன்‌

ஆகுல நீர பிற.

மனம்‌ தூய்மையாக இருப்பதே அறம்‌; மற்றவை ஆரவாரத்தைத்‌ தவிர வேறொன்றுமில்லை.

In spotless mind virtue is found

And not in show and swelling sound.

அழுக்கா றவாவெகுளி இன்னாச்சொல்‌ நான்கும்‌

இழுக்கா இயன்ற தறம்‌.

பொறாமை, பேராசை, பொங்கும்‌ கோபம்‌, புண்படுத்தும்‌ சொல்‌ ஆகிய இந்த நான்கும்‌ அறவழிக்குப்‌



பொருந்தாதவைகளாகும்‌.

Four ills eschew and virtue reach,

Lust, anger, envy, evil-speech.

அன்றறிவாம்‌ என்னா தறஞ்செய்க மற்றது

பொன்றுங்கால்‌ பொன்றாத்‌ தூணை.

பிறகு பார்த்துக்கொள்ளலாம்‌ என்று நாள்‌ கடத்தாமல்‌ அறவழியை மேற்கொண்டால்‌ அது ஒருவர்‌
இறந்தபின்‌ கூட அழியாப்‌ புகழாய்‌ நிலைத்துத்‌ துணை நிற்கும்‌.

Do good enow; defer it not

A deathless aid in death if sought.

அறத்தா றிதுவென வேண்டா சிவிகை

பொறுத்தானோ இர்ந்தான்‌ இடை.

அறவழியில்‌ நடப்பவர்கள்‌ பல்லக்கில்‌ உட்கார்ந்து செல்பவர்களைப்‌ போல வாழ்க்கையில்‌ வரும்‌ இன்ப
துன்பங்கள்‌ இரண்டையும்‌ எளியவாகக்‌ கருதி மகிழ்வுடன்‌ பயணத்தை மேற்கொள்வார்கள்‌. தீய வழிக்குத்‌
தங்களை ஆட்படுத்திக்‌ கொண்டவர்களோ பல்லக்கைத்‌ தூக்கிச்‌ சுமப்பவர்களைப்‌ போல இன்பத்திலும்‌
அமைதி கொள்ளாமல்‌, துன்பத்தையும்‌ தாங்கிக்‌ கொள்ளும்‌ மனப்பக்குவமின்றி வாழ்வையே பெரும்‌
சுமையாகக்‌ கருதுவார்கள்‌.

Like stones that block rebirth and pain

Are doing good and good again.

விழ்நாள்‌ படாஅமை நன்றாற்றின்‌ அஃதொருவன்‌



வாழ்நாள்‌ வழியடைக்கும்‌ கல்‌.

பயனற்றதாக ஒருநாள்கூடக்‌ கழிந்து போகாமல்‌, தொடர்ந்து நற்செயல்களில்‌ ஈ.டுபடுபவருக்கு வாழ்க்கைப்‌
பாதையைச்‌ சீராக்கி அமைத்துத்‌ தரும்‌ கல்லாக அந்த நற்செயல்களே விளங்கும்‌

Like stones that block rebirth and pain

Are doing good and good again..

அறத்தான்‌ வருவதே இன்பமற்‌ றெல்லாம்‌

புறத்த புகழும்‌ இல.

தூய்மையான நெஞ்சுடன்‌ நடத்தும்‌ அறவழி வாழ்க்கையில்‌ வருகின்ற புகழால்‌ ஏற்படுவதே இன்பமாகும்‌.
அதற்கு மாறான வழியில்‌ வருவது புகழும்‌ ஆகாது; இன்பமும்‌ ஆகாது.

Weal flows only from virtue done

The rest is rue and renown gone.

செயற்பால தோரும்‌ அறனே ஒருவற்கு

உயற்பால தோரும்‌ பழி.

பழிக்கத்‌ தக்கவைகளைச்‌ செய்யாமல்‌ பாராட்டத்தக்க அறவழிச்‌ செயல்களில்‌ நாட்டம்‌ கொள்வதே
ஒருவர்க்குப்‌ புகழ்‌ சேர்க்கும்‌.

Worthy act is virtue done

Vice is what we ought to shun.


இல்வாழ்வான்‌ என்பான்‌ இயல்புடைய மூவர்க்கும்‌

நல்லாற்றின்‌ நின்ற தூணை.

பெற்றோர்‌, வாழ்க்கைத்‌ துணை, குழந்தைகள்‌ என இயற்கையாக அமைந்திடும்‌ மூவர்க்கும்‌ துணையாக
இருப்பது இல்லறம்‌ நடத்துவோர்‌ கடமையாகும்‌.

The ideal householder is he

Who aids the natural orders there.

தூறந்தார்க்கும்‌ தூவ்வா தவர்க்கும்‌ இறந்தார்க்கும்‌

இல்வாழ்வான்‌ என்பான்‌ நூணை.

பற்றற்ற துறவிகட்கும்‌, பசியால்‌ வாடுவோர்க்கும்‌, பாதுகாப்பற்றவர்க்கும்‌ இல்லற வாழ்வு நடத்துவோர்‌
துணையாக இருத்தல்‌ வேண்டும்‌.

His help the monk and retired share,

And celibate students are his care.


தென்புலத்தார்‌ தெய்வம்‌ விருந்தொக்கல்‌ தானென்றாங்கு

ஐம்புலத்தா றோம்பல்‌ தலை.

வாழ்ந்து மறைந்தோரை நினைவுகூர்தல்‌, வாழ்வாங்கு வாழ்வோரைப்‌ போற்றுதல்‌, விருந்தோம்பல்‌, சுற்றம்‌
பேணல்‌ ஆகிய கடமைகளை நிறைவேற்றத்‌ தன்னை நிலைப்படுத்திக்‌ கொள்ளல்‌ எனப்படும்‌ ஐவகை
அறநெறிகளும்‌ இல்வாழ்வுக்குரியனவாம்‌.

By dutiful householder's aid

God, manes, kin, self and guests are served.

பழியஞ்சிப்‌ பாத்தாண்‌ உடைத்தாயின்‌ வாழ்க்கை

வழியெஞ்சல்‌ எஞ்ஞான்றும்‌ இல.

பழிக்கு அஞ்சாமல்‌ சேர்த்த பொருள்‌ கணக்கின்றி இருப்பினும்‌ அதைவிட, பழிக்கு அஞ்சிச்‌ சேர்த்த
பொருளைப்‌ பகுத்து உண்ணும்‌ பண்பிலேதான்‌ வாழ்க்கையின்‌ ஒழுக்கமே இருக்கிறது.

Sin he shuns and food he shares

His home is bright and brighter fares.

அன்பும்‌ அறனும்‌ உடைத்தாயின்‌ இல்வாழ்க்கை

பண்பும்‌ பயனும்‌ அது.

இல்வாழ்க்கை பண்புடையதாகவும்‌ பயனுடையதாகவும்‌ விளங்குவதற்கு அன்பான உள்ளமும்‌
அதையொட்டிய நல்ல செயல்களும்‌ தேவை.

In grace and gain the home excels,

Where love with virtue sweetly dwells.

அறத்தாற்றின்‌ இல்வாழ்க்கை ஆற்றின்‌ புறத்தாற்றில்‌

போஒய்ப்‌ பெறுவ தெவன்‌.

அறநெறியில்‌ இல்வாழ்க்கையை அமைத்துக்‌ கொண்டவர்கள்‌ பெற்றிடும்‌ பயனை, வேறு நெறியில்‌ சென்று
பெற்றிட இயலுமோ ? இயலாது.

Who turns from righteous family

To be a monk, what profits he?

இயல்பினான்‌ இல்வாழ்க்கை வாழ்பவன்‌ என்பான்‌

முயல்வாருள்‌ எல்லாம்‌ தலை.

நல்வாழ்வுக்கான முயற்சிகளை மேற்கொள்வோரில்‌ தலையானவராகத்‌ திகழ்பவர்‌, இல்வாழ்வின்‌
இலக்கணமுணர்ந்து அதற்கேற்ப வாழ்பவர்தான்‌.

Of all who strive for bliss, the great

Is he who leads the married state.

ஆற்றின்‌ ஒழுக்கி அறனிழுக்கா இல்வாழ்க்கை

நோற்பாரின்‌ நோன்மை உடைத்து.

தானும்‌ அறவழியில்‌ நடந்து, பிறரையும்‌ அவ்வழியில்‌ நடக்கச்‌ செய்திடுவோரின்‌ இல்வாழ்க்கை, துறவிகள்‌
கடைப்பிடிக்கும்‌ நோன்பைவிடப்‌ பெருமையுடையதாகும்‌.

Straight in virtue, right in living

Make men brighter than monks praying.

அறன்‌ எனப்‌ பட்டதே இல்வாழ்க்கை அஃதும்‌

பிறன்பழிப்ப தில்லாயின்‌ நன்று.

பழிப்புக்கு இடமில்லாத இல்வாழ்க்கை இல்லறம்‌ எனப்‌ போற்றப்படும்‌.

Home-life and virtue, are the same;

Which spotless monkhood too can claim.

வையத்துள்‌ வாழ்வாங்கு வாழ்பவன்‌ வான்‌ உறையும்‌

தெய்வத்துள்‌ வைக்கப்‌ படும்‌.

தெய்வத்துக்கென எத்தனையோ அருங்குணங்கள்‌ கூறப்படுகின்றன. உலகில்‌ வாழ வேண்டிய அறநெறியில்‌
நின்று வாழ்கிறவன்‌ வானில்‌ வாழ்வதாகச்‌ சொல்லப்படும்‌ தெய்வத்துக்கு இணையாக வைத்து
மதிக்கப்படுவான்‌.

He is a man of divine worth

Who lives in ideal home on earth.



மனைத்தக்க மாண்புடையள்‌ ஆகித்தற்‌ கொண்டான்‌
வளத்தக்காள்‌ வாழ்க்கைத்‌ துணை.

இல்லறத்திற்குரிய பண்புகளுடன்‌, பொருள்‌ வளத்துக்குத்‌ தக்கவாறு குடும்பம்‌ நடத்துபவள்‌, கணவனின்‌




வாழ்வுக்குப்‌ பெருந்துணையாவாள்‌.

A good housewife befits the house,

Spending with thrift the mate's resource.

மனைமாட்சி இல்லாள்கண்‌ இல்லாயின்‌ வாழ்க்கை

எனைமாட்சித்‌ தாயினும்‌ இல்‌.

நற்பண்புள்ள மனைவி அமையாத இல்வாழ்க்கை எவ்வளவு சிறப்புடையதாக இருந்தாலும்‌ அதற்குத்‌
தனிச்சிறப்புக்‌ கிடையாது.

Bright is home when wife is chaste.

If not all greatness is but waste.

இல்லதென்‌ இல்லவள்‌ மாண்பானால்‌ உள்ளதென்‌

இல்லவள்‌ மாணாக்‌ கடை.

நல்ல பண்புடைய மனைவி அமைந்த வாழ்க்கையில்‌ எல்லாம்‌ இருக்கும்‌. அப்படியொரு மனைவி அமையாத
வாழ்க்கையில்‌ எதுவுமே இருக்காது.

What is rare when wife is good.

What can be there when she is bad?

பெண்ணின்‌ பெருந்தக்க யாவுள கற்பென்னும்‌

திண்மைஉண்‌ டாகப்‌ பெறின்‌.

கற்பென்னும்‌ திண்மை கொண்ட பெண்மையின்‌ உறுதிப்‌ பண்பைப்‌ பெற்றுவிட்டால்‌, அதைவிடப்‌
பெருமைக்குரியது வேறு யாது?

What greater fortune is for men

Than a constant chaste woman?

தெய்வம்‌ தொழாஅள்‌ கொழுநற்‌ றொழுதெழுவாள்‌

பெய்யெனப்‌ பெய்யும்‌ மழை.

கணவன்‌ வாக்கினைக்‌ கடவுள்‌ வாக்கினை விட மேலானதாகக்‌ கருதி அவனையே தொழுதிடும்‌ மனைவி
பெய்‌ என ஆணையிட்டவுடன்‌ அஞ்சி நடுங்கிப்‌ பெய்கின்ற மழையைப்‌ போலத்‌ தன்னை அடிமையாக
எண்ணிக்‌ கொள்பவளாவாள்‌.

Her spouse before God who adores,

Is like rain that at request pours.

தற்காத்துத்‌ தற்கொண்டாற்‌ பேணித்‌ தகைசான்ற

சொற்காத்துச்‌ சோர்விலாள்‌ பெண்‌.

கற்புநெறியில்‌ தன்னையும்‌ தன்‌ கணவனையும்‌ காத்துக்‌ கொண்டு, தமக்குப்‌ பெருமை சேர்க்கும்‌ புகழையும்‌
காப்பாற்றிக்‌ கொள்வதில்‌ உறுதி குலையாமல்‌ இருப்பவள்‌ பெண்‌.

The good wife guards herself from blame,

She tends her spouse and brings him fame.

சிறைகாக்கும்‌ காப்பெவன்‌ செய்யும்‌ மகளிர்‌

நிறைகாக்கும்‌ காப்பே தலை.

தம்மைத்‌ தாமே காத்துக்கொண்டு சிறந்த பண்புடன்‌ வாழும்‌ மகளிரை அடிமைகளாக நடத்த எண்ணுவது
அறியாமையாகும்‌.

Of what avail are watch and ward?

Their purity is women's guard.

பெற்றாற்‌ பெறின்பெறுவர்‌ பெண்டிர்‌ பெருஞ்சிறப்புப்‌

புத்தேளிர்‌ வாழும்‌ உலகு.

நற்பண்பு பெற்றவனைக்‌ கணவனாகப்‌ பெற்றால்‌, பெண்டிர்க்கு இல்வாழ்க்கையெனும்‌ புதிய உலகம்‌ பெருஞ்‌
சிறப்பாக அமையும்‌.

Women who win their husbands' heart

Shall flourish where the gods resort.

புகழ்புரிந்‌ தில்லிலோர்க்‌ கில்லை இகழ்வார்முன்‌

ஏறுபோல்‌ பீடு நடை.

புகழுக்குரிய இல்வாழ்க்கை அமையாதவர்கள்‌, தம்மைப்‌ பழித்துப்‌ பேசுவோர்‌ முன்பு தலைநிமிர்ந்து நடக்க
முடியாமல்‌ குன்றிப்‌ போய்‌ விடுவார்கள்‌.

A cuckold has not the lion-like gait

Before his detractors aright.

மங்கலம்‌ என்ப மனைமாட்சி மற்றதன்‌

நன்கலம்‌ நன்மக்கட்‌ பேறு.

குடும்பத்தின்‌ பண்பாடுதான்‌ இல்வாழ்க்கையின்‌ சிறப்பு; அதற்கு மேலும்‌ சிறப்பு நல்ல பிள்ளைகளைப்‌

பெற்றிருப்பது.
An honest wife is home's delight
And children good are jewels abright.



61. பெறுமவற்றுள்‌ யாமறிவ தில்லை அறிவறிந்த
மக்கட்பே றல்ல பிற.
அறிவில்‌ சிறந்த நல்ல பிள்ளைகளைவிட இல்வாழ்க்கையில்‌ சிறந்த பேறு வேறு எதுவுமில்லை.
The world no higher bliss bestows
Than children virtuous and wise.
62. எழுபிறப்பும்‌ தீயவை தீண்டா பழிபிறங்காப்‌
பண்புடை மக்கட்‌ பெறின்‌.
பெற்றெடுக்கும்‌ மக்கள்‌ பழிபடராத பண்புடையவர்களாக இருப்பின்‌, ஏழேழு தலைமுறை எனும்‌ அளவுக்குக்‌
காலமெல்லாம்‌ எந்தத்‌ தீமையும்‌ தீண்டாது.
No evil comes and no blemish;
Noble sons bring all we wish.
63. தம்பொருள்‌ என்பதம்‌ மக்கள்‌ அவர்பொருள்‌
தந்தம்‌ வினையான்‌ வரும்‌.
தம்‌ பொருள்‌ என்பது தம்மக்களையேயாம்‌. அம்மக்களின்‌ பொருள்கள்‌ அவரவர்‌ செயல்களின்‌ விளைவாக
வரக்கூடியவை.
Children are one's wealth indeed
Their wealth is measured by their deed.
64. அமிழ்தினும்‌ ஆற்ற இனிதேதம்‌ மக்கள்‌
சிறுகை அளாவிய கூழ்‌.
சிறந்த பொருளை அமிழ்தம்‌ எனக்‌ குறிப்பிட்டாலுங்கூடத்‌ தம்முடைய குழந்தைகளின்‌ பிஞ்சுக்கரத்தால்‌
அளாவப்பட்ட கூழ்‌ அந்த அமிழ்தத்தைவிடச்‌ சுவையானதாகிவிடுகிறது.
The food is more than nectar sweet
In which one's children hands insert.
65. மக்கள்மெய்‌ தீண்டல்‌ உடற்கின்பம்‌ மற்றவர்‌
சொற்கேட்டல்‌ இன்பம்‌ செவிக்கு.
தம்‌ குழந்தைகளைத்‌ தழுவி மகிழ்வது உடலுக்கு இன்பத்தையும்‌, அந்தக்‌ குழந்தைகளின்‌ மழலை மொழி
கேட்பது செவிக்கு இன்பத்தையும்‌ வழங்கும்‌.
Children's touch delights the body
Sweet to ears are their words lovely.
66. ருழலினி தியாழினி தென்பதம்‌ மக்கள்‌
மழலைச்சொல்‌ கேளா தவர்‌.
தங்கள்‌ குழந்தைகளின்‌ மழலைச்‌ சொல்லைக்‌ கேட்காதவர்கள்தான்‌ குழலோசை, யாழோசை ஆகிய
இரண்டும்‌ இனிமையானவை என்று கூறுவார்கள்‌.
The flute and lute are sweet they say
Deaf to baby's babble's lay!
67. தந்‌ைத மகற்காற்றும்‌ நன்றி அவையத்து
முந்தி இருப்பச்‌ செயல்‌.
தந்‌ைத தன்‌ மக்களுக்குச்‌ செய்யவேண்டிய நல்லுதவி அவர்களை அறிஞர்கள்‌ அவையில்‌ புகழுடன்‌
விளங்குமாறு ஆக்குதலே ஆகும்‌.
A father's duty to his son is
To seat him in front of the wise.
68. தம்மின்தம்‌ மக்கள்‌ அறிவுடைமை மாநிலத்து
மன்னுயிர்க்‌ கெல்லாம்‌ இனிது.
பெற்றோரைக்‌ காட்டிலும்‌ பிள்ளைகள்‌ அறிவிற்‌ சிறந்து விளங்கினால்‌, அது பெற்றோருக்கு மட்டுமேயன்றி
உலகில்‌ வாழும்‌ அனைவருக்கும்‌ அக மகிழ்ச்சி தருவதாகும்‌.
With joy the hearts of parents swell
To see their children themselves excel.



ஈன்ற பொழுதின்‌ பெரிதுவக்கும்‌ தன்மகனைச்‌

சான்றோன்‌ எனக்கேட்ட தாய்‌.

நல்ல மகனைப்‌ பெற்றெடுத்தவள்‌ என்று ஊரார்‌ பாராட்டும்‌ பொழுது அவனைப்‌ பெற்றபொழுது அடைந்த
மகிழ்ச்சியைவிட அதிக மகிழ்ச்சியை அந்தத்‌ தாய்‌ அடைவாள்‌.

The mother, hearing her son's merit

Delights more than when she begot.

மகன்‌ தந்தைக்‌ காற்றும்‌ உதவி இவன்தந்தை

என்நோற்றான்‌ கொல்லெனும்‌ சொல்‌.

“ஆகா! இவனைப்‌ பிள்ளையாகப்‌ பெற்றது இவன்‌ தந்‌ைத பெற்ற பெறும்பேறு”', என்று ஒரு மகன்‌
புகழப்படுவதுதான்‌, அவன்‌ தன்னுடைய தந்தைக்குச்‌ செய்யக்கூடிய கைம்மாறு எனப்படும்‌.

The son to sire this word is debt 70

"What penance such a son begot!"

AsTYyMLsww - Loving-Kindness


அன்பிற்கும்‌ உண்டோ அடைக்குந்தாழ்‌ ஆர்வலர்‌

புண்கணீர்‌ பூசல்‌ தரும்‌.

உள்ளத்தில்‌ இருக்கும்‌ அன்பைத்‌ தாழ்ப்பாள்‌ போட்டு அடைத்து வைக்க முடியாது. அன்புக்குரியவரின்‌
துன்பங்காணுமிடத்து, கண்ணீர்த்துளி வாயிலாக அது வெளிப்பட்டுவிடும்‌.

What bolt can bar true love in fact

The tricking tears reveal the heart.

அன்பிலார்‌ எல்லாம்‌ தமக்குரியர்‌ அன்புடையார்‌

என்பும்‌ உரியார்‌ பிறர்க்கு.

அன்பு இல்லாதவர்‌, எல்லாம்‌ தமக்கே என உரிமை கொண்டாடுவர்‌; அன்பு உடையவரோ தம்‌ உடல்‌,
பொருள்‌, ஆவி ஆகிய அனைத்தும்‌ பிறருக்கென எண்ணிடுவர்‌.

To selves belong the loveless ones;

To oth'rs the loving e'en to bones.

அன்போ டியைந்த வழக்கென்ப ஆருயிர்க்கு

என்போ டியைந்த தொடர்பு.

உயிரும்‌ உடலும்போல்‌ அன்பும்‌ செயலும்‌ இணைந்திருப்பதே உயர்ந்த பொருத்தமாகும்‌.

Soul is encased in frame of bone

To taste the life of love alone.

அன்பீனும்‌ ஆர்வம்‌ உடைமை அதுஈ.னும்‌

நண்பென்னும்‌ நாடாச்‌ சிறப்பு.

அன்பு பிறரிடம்‌ பற்றுள்ளம்‌ கொள்ளச்‌ செய்யும்‌ அந்த உள்ளம்‌, நட்பு எனும்‌ பெருஞ்சிறப்பை உருவாக்கும்‌.
Love yields aspiration and thence

Friendship springs up in excellence.

அன்புற்‌ றமர்ந்த வழக்கென்ப வையகத்து

இன்புற்றார்‌ எய்தும்‌ சிறப்பு.

உலகில்‌ இன்புற்று வாழ்கின்றவர்க்கு வாய்க்கும்‌ சிறப்பு, அவர்‌ அன்புள்ளம்‌ கொண்டவராக விளங்குவதன்‌
பயனே என்று கூறலாம்‌.

The crowning joy of home life flows

From peaceful psychic love always.

அறத்திற்கே அன்புசார்‌ பென்ப அறியார்‌

மறத்திற்கும்‌ அஃதே தூணை.

வீரச்‌ செயல்களுக்கும்‌ அன்பு துணையாகத்‌ திகழ்கிறது என்பதை அறியாதவர்களே, அறச்‌ செயல்களுக்கு
மட்டுமே அன்பு துணையாக இருப்பதாகக்‌ கூறுவார்கள்‌.

"Love is virtue's friend" say know-nots

It helps us against evil plots.
என்பி லதனை வெயில்போலக்‌ காயுமே

அன்பி லதனை அறம்‌.

அறம்‌ எதுவென அறிந்தும்‌ அதனைக்‌ கடைப்பிடிக்காதவரை, அவரது மனச்சாட்சியே வாட்டி வதைக்கும்‌.
அது வெயிலின்‌ வெம்மை புழுவை வாட்டுவதுபோல இருக்கும்‌.

Justice burns the loveless form

Like solar blaze the boneless worm.




அன்பகத்‌ தில்லா உயிர்வாழ்க்கை வன்பாற்கண்‌

வற்றல்‌ மரந்தளிர்த்‌ தற்று.

மனத்தில்‌ அன்பு இல்லாதவருடைய வாழ்க்கை, பாலைவனத்தில்‌ பட்டமரம்‌ தளிர்த்தது போன்றது.
Life bereft of love is gloom

Can sapless tree in desert bloom?

புறத்தூறுப்‌ பெல்லாம்‌ எவன்செய்யும்‌ யாக்கை

அகத்துறுப்‌ பன்பி லவர்க்கு.

அன்பு எனும்‌ அகத்து உறுப்பு இல்லாதவர்க்குப்‌ புறத்து உறுப்புகள்‌ அழகாக இருந்து என்ன பயன்‌ ?
Love is the heart which limbs must move,

Or vain the outer parts will prove.

அன்பின்‌ வழிய துயிர்நிலை அஃதிலார்க்கு

என்புதோல்‌ போர்த்த உடம்பு.

அன்புநெஞ்சத்தின்‌ வழியில்‌ இயங்குவதே உயிருள்ள உடலாகும்‌; இல்லையேல்‌, அது எலும்பைத்‌ தோல்‌
போர்த்திய வெறும்‌ உடலேயாகும்‌.

The seat of life is love alone;

Or beings are but skin and bone!



இருந்தோம்பி இல்வாழ்வ தெல்லாம்‌ விருந்தோம்பி

வேளாண்மை செய்தற்‌ பொருட்டு.

இல்லறத்தைப்‌ போற்றி வாழ்வது, விருந்தினரை வரவேற்று, அவர்க்கு வேண்டிய உதவிகளைச்‌
செய்வதற்காகவே.

Men set up home, toil and earn

To tend the guests and do good turn.

விருந்து புறத்ததாத்‌ தானுண்டல்‌ சாவா

மருந்தெனினும்‌ வேண்டற்பாற்‌ றன்று.

விருந்தினராக வந்தவரை வெளியே விட்டுவிட்டுச்‌ சாகாத மருந்தாக இருந்தாலும்‌ அதனைத்‌ தான்‌ மட்டும்‌
உண்பது விரும்பத்‌ தக்க பண்பாடல்ல.

To keep out guests cannot be good

Albeit you eat nector-like food.

வருவிருந்து வைகலும்‌ ஓம்புவான்‌ வாழ்க்கை

பருவந்து பாழ்படுதல்‌ இன்று.

விருந்தினரை நாள்தோறும்‌ வரவேற்று மகிழ்பவரின்‌ வாழ்க்கை, அதன்‌ காரணமாகத்‌ துன்பமுற்றுக்‌
கெட்டொழிவதில்லை.

Who tends his guests day in and out

His life in want never wears out.

அகனமர்ந்து செய்யாள்‌ உறையும்‌ முகனமர்ந்து

நல்விருந்‌ தோம்புவான்‌ இல்‌.

மனமகிழ்ச்சியை முகமலர்ச்சியால்‌ காட்டி விருந்தினரை வரவேற்பவர்‌ வீட்டில்‌ அமர்ந்து செல்வம்‌ எனும்‌
திருமகள்‌ வாழ்வாள்‌.

The goddess of wealth will gladly rest

Where smiles welcome the worthy guest.

வித்தும்‌ இடல் வேண்டும்‌ கொல்லோ விருந்தோம்பி

மிச்சில்‌ மிசைவான்‌ புலம்‌.

விருந்தினர்க்கு முதலில்‌ உணவளித்து மிஞ்சியதை உண்டு வாழும்‌ பண்பாளன்‌, தன்‌ நிலத்திற்குரிய
விதையைக்கூட விருந்தோம்பலுக்குப்‌ பயன்படுத்தாமல்‌ இருப்பானா?

Should his field be sown who first

Feeds the guests and eats the rest?

செல்விருந்‌ தோம்பி வருவிருந்து பார்த்திருப்பான்‌

நல்விருந்து வானத்‌ தவர்க்கு.

வந்த விருந்தினரை உபசரித்து அவர்களை வழியனுப்பி வைக்கும்போதே, மேலும்‌ வரக்கூடிய விருந்தினரை
ஆவலுடன்‌ எதிர்நோக்கி நிற்பவனை, புகழ்வானில்‌ இருப்போர்‌ நல்ல விருந்தினன்‌ என்று வரவேற்றுப்‌
போற்றுவர்‌.



Who tends a guest and looks for next

Is a welcome guest in heaven's feast.

இனைத்துணைத்‌ தென்பதொன்‌ றில்லை விருந்தின்‌

துணைத்தூணை வேள்விப்‌ பயன்‌.

விருந்தினராக வந்தவரின்‌ சிறப்பை எண்ணிப்‌ பார்த்து விருந்தோம்பலை ஒரு வேள்வியாகவே கருதலாம்‌.
Worth of the guest of quality

Is worth of hospitality.

பரிந்தோம்பிப்‌ பற்றற்றேம்‌ என்பர்‌ விருந்தோம்பி

வேள்வி தலைப்படா தார்‌.

செல்வத்தைச்‌ சேர்த்துவைத்து அதனை இழக்கும்போது, விருந்தோம்பல்‌ எனும்‌ வேள்விக்கு அது
பயன்படுத்தப்படாமற்‌ போயிற்றே என வருந்துவார்கள்‌.

Who loathe guest-service one day cry:

"We toil and store; but life is dry".

உடைமையுள்‌ இன்மை விருந்தோம்பல்‌ ஓம்பா

மடமை மடவார்கண்‌ உண்டு.

விருந்தினரை வரவேற்றுப்‌ போற்றத்‌ தெரியாத அறிவற்றவர்கள்‌ எவ்வளவு பணம்‌ படைத்தவர்களாக
இருந்தாலும்‌ தரித்திரம்‌ பிடித்தவர்களாகவே கருதப்படுவார்கள்‌.

The man of wealth is poor indeed

Whose folly fails the guest to feed.

மோப்பக்‌ குழையும்‌ அனிச்சம்‌ முகந்திரிந்து

நோக்கக்‌ குழையும்‌ விருந்து.

அனிச்சம்‌ எனப்படும்‌ பூ, முகர்ந்தவுடன்‌ வாடி விடக்‌ கூடியது. அதுபோல்‌ சற்று முகங்கோணி வரவேற்றாலே
விருந்தினர்‌ வாடிவிடுவர்‌.

Anicham smelt withers: like that

A wry-faced look withers the guest.





இன்சொலால்‌ ஈ.ரம்‌ அளைஇப்‌ படிறிலவாம்‌

செம்பொருள்‌ கண்டார்வாய்ச்‌ சொல்‌.

ஒருவர்‌ வாயிலிருந்து வரும்‌ சொல்‌ அன்பு கலந்ததாகவும்‌, வஞ்சனையற்றதாகவும்‌, வாய்மையுடையதாகவும்‌
இருப்பின்‌ அதுவே இன்சொல்‌ எனப்படும்‌.

The words of Seers are lovely sweet

Merciful and free from deceit.

அகன்்‌அமர்ந்‌ தீதலின்‌ நன்றே முகனமர்ந்து

இன்சொலன்‌ ஆகப்‌ பெறின்‌.

முகம்‌ மலர்ந்து இனிமையாகப்‌ பேசுவது, அகம்‌ குளிர்ந்து ஒன்றைக்‌ கொடுப்பதை விட மேலான பண்பாகும்‌.
Sweet words from smiling lips dispense

More joys than heart's beneficence.

முகத்தான்‌ அமர்ந்தின்நு நோக்கி அகத்தானாம்‌

இன்சொ லினதே அறம்‌.

முகம்‌ மலர நோக்கி, அகம்‌ மலர இனிய சொற்களைக்‌ கூறுவதே அறவழியில்‌ அமைந்த பண்பாகும்‌.
Calm face, sweet look, kind words from heart

Such is the gracious virtue's part.

தூன்புறாஉம்‌ துவ்வாமை இல்லாகும்‌ யார்மாட்டும்‌

இன்புறாஉம்‌ இன்சொ லவர்க்கு.

இன்சொல்‌ பேசி எல்லோரிடத்திலும்‌ கனிவுடன்‌ பழகுவோர்க்கு “நட்பில்‌ வறுமை' எனும்‌ துன்பமில்லை.
Whose loving words delight each one

The woe of want from them is gone.

பணிவுடையன்‌ இன்சொலன்‌ ஆதல்‌ ஒருவற்கு

அணியல்ல மற்றுப்‌ பிற.

அடக்கமான பண்பும்‌, இனிமையாகப்‌ பேசும்‌ இயல்பும்‌ தவிர, ஒருவருக்குச்‌ சிறந்த அணிகலன்‌ வேறு
இருக்க முடியாது.

To be humble and sweet words speak

No other jewel do wise men seek.

96. அல்லவை தேய அறம்பெருரும்‌ நல்லவை
நாடி இனிய சொலின்‌.
தீய செயல்களை அகற்றி அறநெறி தழைக்கச்‌ செய்ய வேண்டுமானால்‌, இனிய சொற்களைப்‌ பயன்படுத்தி
நல்வழி எதுவெனக்‌ காட்ட வேண்டும்‌.
His sins vanish, his virtues grow
Whose fruitful words with sweetness flow.
97. நயன்்‌ஈ.ன்று நன்றி பயக்கும்‌ பயன்‌ ஈன்று
பண்பின்‌ தலைப்பிரியாச்‌ சொல்‌.
நன்மையான பயனைத்‌ தரக்கூடிய நல்ல பண்பிலிருந்து விலகாத சொற்கள்‌ அவற்றைக்‌ கூறுவோருக்கும்‌
இன்பத்தையும்‌, நன்மையையும்‌ உண்டாக்கக்‌ கூடியவைகளாகும்‌.
The fruitful courteous kindly words
Lead to goodness and graceful deeds.
98. சிறுமையுள்‌ நீங்கிய இன்சொல்‌ மறுமையும்‌
இம்மையும்‌ இன்பம்‌ தரும்‌.
சிறுமைத்தனமற்ற இனியசொல்‌ ஒருவனுக்கு அவன்‌ வாழும்‌ போதும்‌, வாழ்ந்து மறைந்த பிறகும்‌ புகழைத்‌
தரக்கூடியதாகும்‌.
Kind words free from meanness delight
This life on earth and life the next.
99. இன்சொல்‌ இனிதீன்றல்‌ காண்பான்‌ எவன்கொலோ
வன்சொல்‌ வழங்கு வது.
இனிய சொற்கள்‌ இன்பத்தை வழங்கும்‌ என்பதை உணர்ந்தவர்‌ அதற்கு மாறாக எதற்காகக்‌ கடுஞ்‌
சொற்களைப்‌ பயன்படுத்த வேண்டும்‌ ?
Who sees the sweets of sweetness here
To use harsh words how can he dare?
100. இனிய உளவாக இன்னாத கூறல்‌
கனிஇருப்பக்‌ காய்கவர்ந்‌ தற்று.
இனிமையான சொற்கள்‌ இருக்கும்போது அவற்றை விடுத்துக்‌ கடுமையாகப்‌ பேசுவது கனிகளை ஒதுக்கி
விட்டுக்‌ காய்களைப்‌ பறித்துத்‌ தின்பதற்குச்‌ சமமாகும்‌.
Leaving ripe fruits the raw he eats
Who speaks harsh words when sweet word suits.

"""

# Regular expression compilation
pattern = re.compile(r'(.*?\n.*?\n)(.*?\n)(.*?)', re.DOTALL)

# Search for matches
matches = pattern.findall(text)

# Process matches
for match in matches:
    print("Verse in Tamil:", match[0].strip())
    print("Meaning in Tamil:", match[1].strip())
    print("Translation in English:", match[2].strip())
    print("\n---\n")
```
